```python
import pandas as pd
import numpy as np
import glob

from tqdm import tqdm
from easydict import EasyDict
```

## Feature extraction
- FE í´ë”ì— ìˆëŠ” csvë¥¼ load í•˜ì—¬ ê° featureë¥¼ ì¶”ì¶œ

```python
features = EasyDict()

for file in tqdm(glob.glob("/opt/ml/input/data/FE/*.csv")):
        #get name
        name = file.split('/')[-1].split('.')[0] # get name
        df = pd.read_csv(file)

        #dataframeì—ì„œ í•„ìš”í•œ feature ë§Œ ì¶”ì¶œ
        features[name] = df[name]
```

```python
# Save csv
raw_data = pd.read_csv("/opt/ml/input/data/FE_total_data.csv")
```

## ğŸ›¢ëª¨ë“  featureë¥¼ ì‚¬ìš©í•˜ê³  ì‹¶ì„ ë•Œ ì‚¬ìš©
